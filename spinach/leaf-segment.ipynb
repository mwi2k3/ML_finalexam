```python
data_yaml_content = """
path: /kaggle/working/leaf_dataset_grouped_31_05
train: images/train
val: images/val
test: images/test
nc: 1
names: ['leaf']
"""

# Ghi vào file
with open('/kaggle/working/leaf_dataset_grouped_31_05/data.yaml', 'w') as f:
    f.write(data_yaml_content.strip())

print("✅ Đã tạo file data.yaml thành công.")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n-seg.pt")  # hoặc yolov8m-seg.pt, yolov8s-seg.pt tùy bạn

model.train(
    data="data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    name="segmentation_run"
)
```
